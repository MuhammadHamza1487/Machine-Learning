```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
vals = np.linspace(0, 10, 100)

sin_values = np.sin(vals)
cos_values = np.cos(vals)
```

```python
vals
```

# Plotting the Data

```python
# Sin wave
plt.plot(vals, sin_values)
plt.show()
```

```python
# Cosine Wave
plt.plot(vals, cos_values)
plt.ylabel('cosine value')
plt.xlabel('angle')
plt.title('cosine wave')
plt.show()
```

```python
plt.plot(vals, sin_values, 'r+')
plt.plot(vals, cos_values, 'gx')
# g. | g- | 
plt.show()
```

# Bar Plot

```python
fig = plt.figure()
ax =  fig.add_axes([0, 0, 1, 1])
languages = ['English', 'German', 'French', 'Arabic', 'Italian']
people = [100, 34, 2, 67, 20]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('No of people')
plt.show()
```

# Pie Charts

```python
fig = plt.figure()
ax =  fig.add_axes([0, 0, 1, 1])
languages = ['English', 'German', 'French', 'Arabic', 'Italian']
people = [100, 34, 12, 67, 20]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

# Scattering



```python
fig = plt.figure()
ax =  fig.add_axes([0, 0, 1, 1])
ax.scatter(vals, sin_values, color='r')
ax.scatter(vals, cos_values, color='g')
plt.show()
```

```python
# 3-D scatter plot
ax = plt.axes(projection='3d')
random_arr = 20*np.random.rand(100)
a = np.sin(random_arr)
b = np.cos(random_arr)

ax.scatter(a, b, random_arr)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 3D surface plot
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

# Subplotting

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.plot(vals, sin_values)

plt.subplot(1, 2, 2)
plt.plot(vals, cos_values)

plt.tight_layout  # Adjust spacing to prevent overlap
plt.show()
```
