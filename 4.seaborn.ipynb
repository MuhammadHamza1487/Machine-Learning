```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
import numpy as np
```

Seaborn has some builtin datasets

```python
# total bill vs tip dataset
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
tips_df.shape
```

```python
sns.set_theme()
```

```python
sns.relplot(data=tips_df, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

# IRIS Dataset

```python
iris_df = sns.load_dataset('iris')
iris_df.head()
```

```python
sns.scatterplot(data=iris_df, x='sepal_length', y='petal_width', hue='species')
```

# Titanic Dataset

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
titanic_df.shape
```

### Count Plot

```python
sns.countplot(x='pclass', data=titanic_df)
```

# Bar Plot

```python
sns.barplot(x='sex', y='survived', data=titanic_df)
```
